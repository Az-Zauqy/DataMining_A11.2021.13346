```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data =pd.read_csv("Data.csv")
data.head()
```

```python
data.info()
```

```python
data['age_group'].unique()
```

```python
age_group = '0'
data[data['age_group']==age_group]
```

```python
# konversi kolom 'age_group' ke tipe data integer
data['age_group'] = data['age_group'].astype('Int64')

# buat kondisi untuk memfilter nilai integer positif
age_cond = (data['age_group'].notnull()) & (data['age_group'] > 0)

# gunakan kondisi untuk memfilter data
unique_countries = data[age_cond]['Standard population Germany 2011'].unique()
unique_countries.sort()
unique_countries
```

```python
country_confirmed_case = []
for i in unique_countries:
  country_confirmed_case.append(data[data.age_group>0][data['Standard population Germany 2011']==i][data['age_group']==age_group].age_group.sum())
```

```python
print(set(zip(unique_countries, country_confirmed_case)))
```

```python
unique_populasi = data['Standard population Germany 1987'][data.age_group>1].unique()
unique_populasi
```

```python
populasi_confirmed_case = []
for i in unique_populasi:
  populasi_confirmed_case.append(data[data.age_group>0][data['Standard population Germany 1987']==i][data['age_group']==age_group].age_group.sum())
```

```python
nan_indices = []
for i in range(len(unique_populasi)):
  if type(unique_populasi[i])==float:
    nan_indices.append(i)
```

```python
unique_populasi = list(unique_populasi)
province_confirmed_case = list(populasi_confirmed_case)
```

```python
for i in nan_indices:
  unique_populasi.pop(i)
  province_confirmed_case.pop(i)
```

```python
unique_populasi
```

```python
plt.barh(unique_countries, country_confirmed_case)
plt.show()
```

```python
plt.figure(figsize=(50,40))
plt.barh(unique_populasi, populasi_confirmed_case)
plt.show()
```

```python
import numpy as np
filtered_countries = []
filtered_cases = []
for i in range(len(unique_countries)):
    if not np.isnan(country_confirmed_case[i]):
        filtered_countries.append(unique_countries[i])
        filtered_cases.append(country_confirmed_case[i])
plt.pie(filtered_cases, labels=filtered_countries, autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(20,20))
plt.pie(populasi_confirmed_case)
plt.legend(unique_populasi, loc='best')
plt.show()
```
